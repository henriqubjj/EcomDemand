```python
import os
import glob
from datetime import datetime, timedelta

# Caminho da pasta onde estão os arquivos
folder_path = 'data'

# Data inicial (dia 01/02/2024)
start_date = datetime.strptime("01/02/2024", "%d/%m/%Y")

# Obter todos os arquivos .xls na pasta 'data'
files = glob.glob(os.path.join(folder_path, "*.xls"))

# Ordenar os arquivos pela data de criação
files_sorted = sorted(files, key=lambda x: os.path.getctime(x))

# Reverter a lista para que a ordem fique conforme o Windows (primeiro = 01/02/2024)
files_sorted = list(reversed(files_sorted))

print(f"Total de arquivos encontrados: {len(files_sorted)}")

# Renomear os arquivos sequencialmente, atribuindo datas crescentes
for i, filepath in enumerate(files_sorted):
    new_date = start_date + timedelta(days=i)
    new_filename = "data-" + new_date.strftime("%d-%m-%Y") + ".xls"
    new_filepath = os.path.join(folder_path, new_filename)
    
    os.rename(filepath, new_filepath)
    print(f"Renomeado: {os.path.basename(filepath)} -> {new_filename}")
```
